```python
class Car:

    def __init__(self, car_name, owner, price):
        self.car_name = car_name
        self.owner = owner
        self.price = price
        print("Constructor Called")

    def display(self):
        print("Car Name:", self.car_name)
        print("Owner:", self.owner)
        print("Price:", self.price)

    def __del__(self):
        print("Car object deleted")


# Create object
car_shop = Car("BMW", "Karthik", 4000)

# Display details
car_shop.display()

# Delete object
del car_shop
```

```python
class Animal:
    def __init__(self, name):
        self.name = name
 
    def info(self):
        print("Animal name:", self.name)
 
class Dog(Animal):
    def sound(self):
        print(self.name, "barks")
 
d = Dog("Buddy")
# Inherited method
d.info()    
d.sound()
 
```

```python
class Person:

    def person_details(self):
        print("Name: Karthik")


class Employee(Person):

    def employee_details(self):
        print("Employee ID: 101")


class Manager(Employee):

    def manager_details(self):
        print("Department: AI")


manager = Manager()

manager.person_details()
manager.employee_details()
manager.manager_details()
```

```python
#polymorphism
class Dog:

    def sound(self):
        print("Bark")


class Cat:

    def sound(self):
        print("Meow")


class Cow:

    def sound(self):
        print("Moo")


animals = [Dog(), Cat(), Cow()]

for animal in animals:
    animal.sound()
```

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):

    @abstractmethod
    def start(self):
        pass


class Car(Vehicle):

    def start(self):
        print("Car Started")


car = Car()
car.start()
```

```python
from abc import ABC, abstractmethod

# Abstract Class
class Employee(ABC):

    @abstractmethod
    def work(self):
        pass


# Child Class 1
class Developer(Employee):

    def work(self):
        print("Developer writes code.")


# Child Class 2
class Tester(Employee):

    def work(self):
        print("Tester tests the application.")


# Creating Objects
developer = Developer()
tester = Tester()

# Calling Methods
developer.work()
tester.work()
```
