```python
import matplotlib.pyplot as plt

months = [1,2,3,4,5,6]

sales = [200,250,220,300,350,400]

plt.figure(figsize=(8,5))

plt.plot(
    months,
    sales,
    color="green",
    linewidth=2,
    linestyle="-",
    marker="o"
)

plt.title("Half Year Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid()

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.figure(figsize=(8, 6))
plt.plot(x, y, marker='o', linestyle='-')

for xi, yi in zip(x, y):
    plt.annotate(f'({xi}, {yi})',
                 (xi, yi),
                 textcoords="offset points",
                 xytext=(0, 10),
                 ha='center')

plt.title('Line Chart with Annotations')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
#line chart
import matplotlib.pyplot as plt

months = ["Jan","Feb","Mar","Apr","May"]

product_A = [100,120,150,170,200]
product_B = [90,140,130,180,210]

plt.plot(months, product_A, label="Product A")
plt.plot(months, product_B, label="Product B")

plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

months = ["Jan","Feb","Mar","Apr","May"]

product_A = [100,120,150,170,200]
product_B = [90,140,130,180,210]

plt.plot(
    months,
    product_A,
    color="blue",
    marker="o",
    linewidth=2,
    label="Product A"
)

plt.plot(
    months,
    product_B,
    color="red",
    marker="s",
    linewidth=2,
    label="Product B"
)
plt.title("Sales Comparison")

plt.xlabel("Month")
plt.ylabel("Sales")

plt.grid()

plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

months = ["Jan","Feb","Mar","Apr","May","Jun"]

sales_2025 = [120,150,180,170,220,240]

sales_2026 = [140,160,200,210,250,280]

plt.figure(figsize=(8,5))

plt.plot(
    months,
    sales_2025,
    color="blue",
    marker="o",
    label="2025"
)

plt.plot(
    months,
    sales_2026,
    color="green",
    marker="s",
    label="2026"
)

plt.title("Sales Comparison")

plt.xlabel("Month")
plt.ylabel("Sales")

plt.grid()

plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

students = ["Alice", "Bob", "Charlie", "David"]

marks = [85, 92, 78, 95]

plt.bar(students, marks)

plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.show()
```

```python
import matplotlib.pyplot as plt

students = ["Alice","Bob","Charlie","David"]

marks = [85,92,78,95]

bars = plt.bar(students, marks)

for bar in bars:
    y = bar.get_height()

    plt.text(
        bar.get_x() + bar.get_width()/2,
        y,
        y,
        ha="center"
    )

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

students = ["A","B","C","D"]

class_A = [80,85,90,88]

class_B = [75,82,95,91]

x = np.arange(len(students))

width = 0.35

plt.bar(
    x-width/2,
    class_A,
    width,
    label="Class A"
)

plt.bar(
    x+width/2,
    class_B,
    width,
    label="Class B"
)

plt.xticks(x, students)

plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

products = [
    "Laptop",
    "Phone",
    "Tablet",
    "Monitor"
]

sales = [150,250,100,180]

plt.figure(figsize=(8,5))

plt.bar(
    products,
    sales,
    color="orange",
    edgecolor="black"
)

plt.title("Product Sales")

plt.xlabel("Products")

plt.ylabel("Units Sold")

plt.grid(axis="y")

plt.show()
```

```python
#piechart
import matplotlib.pyplot as plt

products = ["Laptop", "Phone", "Tablet", "Monitor"]

sales = [40, 30, 20, 10]

plt.pie(
    sales,
    labels=products,
    autopct="%1.1f%%"
)

plt.show()
```

```python
import matplotlib.pyplot as plt

products = [
    "Laptop",
    "Phone",
    "Tablet",
    "Monitor"
]

sales = [40,30,20,10]

colors = [
    "skyblue",
    "orange",
    "green",
    "pink"
]

explode = [0.1,0,0,0]

plt.figure(figsize=(6,6))

plt.pie(
    sales,
    labels=products,
    colors=colors,
    autopct="%1.1f%%",
    explode=explode,
    startangle=90,
    shadow=True
)

plt.title("Product Sales Distribution")

plt.show()
```

```python
#Basic Histogram
import matplotlib.pyplot as plt

marks = [
    45,50,52,55,
    60,62,65,68,
    70,72,75,78,
    80,82,85,88,90
]

plt.figure(figsize=(8,5))

plt.hist(
    marks,
    bins=5,
    color="skyblue",
    edgecolor="black"
)

plt.title("Distribution of Student Marks")

plt.xlabel("Marks")

plt.ylabel("Number of Students")

plt.grid(axis="y")

plt.show()
```

```python
import matplotlib.pyplot as plt

salary = [
    25000,28000,30000,32000,
    35000,38000,40000,42000,
    45000,48000,50000
]

plt.hist(
    salary,
    bins=4,
    color="green",
    edgecolor="black"
)

plt.title("Salary Distribution")

plt.xlabel("Salary")

plt.ylabel("Employees")

plt.show()
```

```python
#Scatter plots
import matplotlib.pyplot as plt

hours = [1,2,3,4,5,6,7]

marks = [35,45,55,65,70,85,95]

plt.figure(figsize=(8,5))

plt.scatter(
    hours,
    marks,
    color="green",
    s=120,
    marker="o"
)

plt.title("Hours Studied vs Marks")

plt.xlabel("Hours Studied")

plt.ylabel("Marks")

plt.grid()

plt.show()
```

```python
import matplotlib.pyplot as plt

height = [150,155,160,165,170,175,180]

weight = [45,50,55,60,68,75,82]

plt.scatter(
    height,
    weight,
    color="blue",
    s=100
)

plt.title("Height vs Weight")

plt.xlabel("Height (cm)")

plt.ylabel("Weight (kg)")

plt.grid()

plt.show()
```

```python
import matplotlib.pyplot as plt

experience = [1,2,3,4,5,6,7,8]

salary = [25000,30000,35000,42000,50000,60000,70000,85000]

plt.figure(figsize=(8,5))

plt.scatter(
    experience,
    salary,
    color="orange",
    s=100,
    marker="D"
)

plt.title("Experience vs Salary")

plt.xlabel("Years of Experience")

plt.ylabel("Salary")

plt.grid()

plt.show()
```
