```python
#To find average of 3 numbers
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
num3 = int(input("Enter third number: "))

average = (num1 + num2 + num3) / 3
print("The average of", num1, ",", num2, "and", num3, "is:", average)
```

```python
# conditional statements
#if statement
age = 20

if age >= 18:
    print("Eligible")

#if-else statement
age = 15

if age >= 18:
    print("Adult")
else:
    print("Minor")
```

```python
#if elif else statement
marks = 75

if marks >= 90:
    print("A Grade")

elif marks >= 75:
    print("B Grade")

else:
    print("C Grade")
```

```python
#Loops
#for loop
for i in range(5):
    print(i)
```

```python
#while loop
i = 1

while i <= 5:
    print(i)
    i += 1
```

```python
for i in range(5, 0, -1):
    for j in range(i):
        print(0, end=' ')
    print()
```

```python
#break
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
#con
for i in range(10):
    if i == 5:
        continue
    print(i)
```

```python
#functions
def greet(): #def is used to create a function.
    print("Hello, Welcome to Python!")

greet()# it is the function name.
```

```python
def fruits(*items): #*args : Whatever extra values you pass are collected into one tuple.
    print(items)

fruits("Apple", "Mango", "Orange")
```

```python
#**kwargs : Whatever extra values you pass are collected into one dictionary.
def student(**details): 
    print(details)

student(name="Prashant", age=22, city="Bangalore")
```

```python
#A lambda function is a small anonymous function used for short operations. It is useful when you need a function only once.
square = lambda x: x * x 


print(square(5))
```
